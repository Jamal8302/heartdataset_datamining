## Confusion Matrix

```python
# Print the Confusion Matrix with k =3 and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print('Confusion matrix\n\n', cm)

print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
# The confusion matrix shows 15 + 20 = 35 correct predictions and 17 + 9 = 26 incorrect predictions.

# In this case, we have

# True Positives (Actual Positive:1 and Predict Positive:1) - 15
# True Negatives (Actual Negative:0 and Predict Negative:0) - 20
# False Positives (Actual Negative:0 but Predict Positive:1) - 17 (Type I error)
# False Negatives (Actual Positive:1 but Predict Negative:0) - 9 (Type II error)
```

```python
# Print the Confusion Matrix with k =7 and slice it into four pieces

cm_7 = confusion_matrix(y_test, y_pred_7)

print('Confusion matrix\n\n', cm_7)

print('\nTrue Positives(TP) = ', cm_7[0,0])

print('\nTrue Negatives(TN) = ', cm_7[1,1])

print('\nFalse Positives(FP) = ', cm_7[0,1])

print('\nFalse Negatives(FN) = ', cm_7[1,0])
```

```python
# The above confusion matrix shows 19 + 23 = 42 correct predictions and 13 + 6 = 19 incorrect predictions.

# In this case, we have

# True Positives (Actual Positive:1 and Predict Positive:1) - 83
# True Negatives (Actual Negative:0 and Predict Negative:0) - 54
# False Positives (Actual Negative:0 but Predict Positive:1) - 2 (Type I error)
# False Negatives (Actual Positive:1 but Predict Negative:0) - 1 (Type II error)
```

## Classification metrices

```python
# Classifation report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_7))
```

```python
# Classification accuracy
TP = cm_7[0,0]
TN = cm_7[1,1]
FP = cm_7[0,1]
FN = cm_7[1,0]
```

```python
# print classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Classification accuracy : {0:0.4f}'.format(classification_accuracy))
```

```python
# Classification error

# print classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.4f}'.format(classification_error))
```

```python
# Precision

# print precision score

precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
```

```python
# Recall

recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```

```python
# True Positive Rate
true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.4f}'.format(true_positive_rate))
```

```python
# False Positive Rate
false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.4f}'.format(false_positive_rate))
```

```python
# Specificity
specificity = TN / (TN + FP)

print('Specificity : {0:0.4f}'.format(specificity))
```
